Equalize image:
$f_{eq}(a) = [H(a)*\frac{K-1}{HM}]$

```python
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
```

```python
src = mpimg.imread('images/base_images/4.2.07.tiff', 0)
plt.imshow(src)
```

```python
src.shape
```

```python
src
```

```python
bnw = np.zeros_like(src)
for i, row in enumerate(src):
    for j, pixel in enumerate(row):
        avg = 0
        for channel in pixel:
            avg += channel
        bnw[i][j] = np.ceil(avg/3)
```

```python
plt.imshow(bnw)
```
